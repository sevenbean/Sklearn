```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
X,Y=make_blobs(n_samples=500,n_features=2,random_state=1)
X.shape
```

```python
X
```

```python
#运用Kmeans算法进行分类,n_clusters设置为3
clusters_3=KMeans(n_clusters=3,random_state=0).fit(X)
y_pred=clusters_3.labels_
```

```python
y_pred
```

```python
#对n_cluster=3，使用轮廓系数的高低来对比模型的好坏，轮廓系数越靠近1，越好
from sklearn.metrics import silhouette_score
from sklearn.metrics import silhouette_samples
score_3=silhouette_score(X,y_pred)
score_3
```

```python
clusters_4=KMeans(n_clusters=4,random_state=0).fit(X)
y_pred_4=clusters_4.labels_
```

```python
score_4=silhouette_score(X,y_pred_4)
score_4
```

```python
#返回每个样本自己的轮廓系数
silhouette_samples(X,y_pred_4)
```

```python
#第二种，衡量算法的指标(卡林斯基-哈拉巴斯指数)，卡林斯基-哈拉巴斯指数越高越好
from sklearn.metrics import calinski_harabasz_score
calinski_harabasz_score(X,y_pred)
```
