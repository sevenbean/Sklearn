```python
#包装法
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
data=pd.read_csv("./day08_digit recognizor.csv")
data.head()
```

```python
X=data.iloc[:,1:]
Y=data.iloc[:,0]
```

```python
RFC_=RandomForestClassifier(n_estimators=10,random_state=0)
selector=RFE(RFC_,n_features_to_select=340,step=50).fit(X,Y)
```

```python
'''
返回所有的特征的是否最后被选
中的布尔矩阵，以及.ranking_返回特征的按数次迭代中综合重要性的排名。
'''
selector.support_.sum()
```

```python
selector.ranking_
```

```python
x_warpper=selector.transform(X)
x_warpper.shape
```

```python
from sklearn.feature_selection import RFECV
```

```python
selector=RFECV(RFC_,min_features_to_select=50,cv=10,step=50).fit(X,Y)
selector.support_
```
