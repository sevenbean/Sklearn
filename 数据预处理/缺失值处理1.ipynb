```python
from sklearn.impute import SimpleImputer
import pandas as pd
data=pd.read_csv("./day08_Narrativedata.csv")
data.info()
```

```python
import numpy as np
Age=data.loc[:,"Age"].values
Age=Age[:,np.newaxis]
Age.shape
Age
```

```python
#用0进行填充
impu_0=SimpleImputer(missing_values=np.nan,strategy="constant",fill_value=0)
#用均值进行填充
impu_mean=SimpleImputer(missing_values=np.nan,strategy="mean")
#用中位数填充
impu_median=SimpleImputer(missing_values=np.nan,strategy="median")
impu_0=impu_0.fit_transform(Age)
impu_mean=impu_mean.fit_transform(Age)
impu_median=impu_median.fit_transform(Age)
```

```python
impu_median=SimpleImputer(missing_values=np.nan,strategy="median")
data.loc[:,"Age"]=impu_median.fit_transform(Age)
```

```python
# strategy中表示的是用那个众数进行填充，可以用于非数值型的数据
impu_most_frequent=SimpleImputer(missing_values=np.nan
                                  ,strategy="most_frequent")

Embarked=data.loc[:,"Embarked"].values.reshape(-1,1)
data.loc[:,"Embarked"]=impu_most_frequent.fit_transform(Embarked)
data.info()
```
