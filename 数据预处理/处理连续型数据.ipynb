```python
#Tsai Tsai
'''
在建模中，需要对连续变量离散化，特征离散化后，模型会更稳定，降低了模型过拟合的风险
处理连续型特征：二值化与分段
sklearn.preprocessing.Binarizer
根据阈值将数据二值化（将特征值设置为0或1），用于处理连续型变量。大于阈值的值映射为1，而小于或等于阈
值的值映射为0。默认阈值为0时，特征中所有的正值都映射到1。二值化是对文本计数数据的常见操作，分析人员
可以决定仅考虑某种现象的存在与否。它还可以用作考虑布尔随机变量的估计器的预处理步骤（例如，使用贝叶斯设置中的伯努利分布建模）
'''
from sklearn.preprocessing import Binarizer
import pandas as pd
data=pd.read_csv("./day08_Narrativedata.csv")
data.drop("Unnamed: 0",axis=1,inplace=True)
```

```python
data
```

```python
data.loc[:,"Age"]=data.loc[:,"Age"].fillna(data.loc[:,"Age"].median())
data
```

```python
data.dropna(inplace=True)
```

```python
data.loc[:,"Survived"].unique()
```

```python
#对Survived标签进行处理
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data.loc[:,"Survived"]=le.fit_transform(data.loc[:,"Survived"].values.reshape(-1,1))
le.classes_
```

```python
data.info()
```

```python
#对标签里面的非数值数据进行处理,由于舱门和性别都是名义变量，所以使用独热编码
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
ohe=OneHotEncoder(categories="auto")
results=ohe.fit_transform(data.iloc[:,1:-1]).toarray()
data=pd.concat([data,pd.DataFrame(results)],axis=1)
```

```python
data.drop(["Sex","Embarked"],axis=1,inplace=True)
```

```python
data.dropna(inplace=True)
```

```python
ohe.get_feature_names()
```

```python
data.columns=["Age","Survived",'x0_female', 'x0_male', 'x1_C', 'x1_Q', 'x1_S']
data
```

```python
#开始对连续的数据特征进行二值化
data_2=data.copy()
```

```python
data_2
```

```python
#对年龄进行二值化，大于阈值则设置为1，小于阈值则设置
br=Binarizer(threshold=30)
age=br.fit_transform(data_2.loc[:,"Age"].values.reshape(-1,1))
age
```

```python
from sklearn.preprocessing import KBinsDiscretizer
data_3=data.copy()
```

```python
data_3
```

```python
data_3.loc[:,"Age"]
```

```python
#uniform表示的是等宽分箱即(最大值-最小值)/分箱数
k_bins=KBinsDiscretizer(n_bins=3,encode="ordinal",strategy="uniform")
data_3.loc[:,"Age"]=k_bins.fit_transform(data_3.loc[:,"Age"].values.reshape(-1,1))
data_3.loc[:,"Age"]
```

```python
#等频分箱
data_4=data.copy()
k_bins_quantile=KBinsDiscretizer(n_bins=3,encode="ordinal",strategy="quantile")
data_4.loc[:,"Age"]=k_bins_quantile.fit_transform(data_4.loc[:,"Age"].values.reshape(-1,1))
data_4.loc[:,"Age"]
```

```python
#kmeans分箱
data_5=data.copy()
k_bins_quantile=KBinsDiscretizer(n_bins=3,encode="ordinal",strategy="kmeans")
data_5.loc[:,"Age"]=k_bins_quantile.fit_transform(data_5.loc[:,"Age"].values.reshape(-1,1))
data_5.loc[:,"Age"]
```
