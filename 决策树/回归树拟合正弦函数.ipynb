```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
```

```python
nrg=np.random.RandomState(1)
```

```python
#ravek进行降维
x=np.sort(nrg.rand(80,1)*5,axis=0)
y=np.sin(x).ravel()
y.shape
```

```python
plt.scatter(x,y)
plt.show()
```

```python
#给y标签添加噪声
y[::5]+=3*(0.5-nrg.rand(16))
```

```python
y[::5].shape
```

```python
plt.scatter(x,y)
plt.show()
```

```python
reg_1=DecisionTreeRegressor(max_depth=2)
reg_2=DecisionTreeRegressor(max_depth=5)
reg_1=reg_1.fit(x,y)
reg_2=reg_2.fit(x,y)
```

```python
#升维方法一：
# l=np.arange(1,10,0.1,).reshape(-1,1)
#升维方法二：np.newaxis在哪个位置上，那么他的维度就会在这个地方增加一个维度
l=np.arange(1,10,0.1)
l=l[:,np.newaxis]
l.shape
```

```python
a=np.arange(1,10,0.1)
a=a[np.newaxis,:]
a.shape
```

```python
x_test=np.arange(0.0,5.0,0.01)[:,np.newaxis]
predict_1=reg_1.predict(x_test)
predict_2=reg_2.predict(x_test)
```

```python
predict_1
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x,y,s=20,edgecolors="black",c="darkorange",label="data")
plt.plot(x_test
         ,predict_1
         ,color="cornflowerblue"
         ,label="max_depth=2"
         ,linewidth=2)
plt.plot(x_test
         ,predict_2
         ,color="yellowgreen"
         ,label="max_depth=5"
         ,linewidth=2)
plt.xlabel("data")
plt.ylabel("target")
plt.title("Decision Tree Regression")
plt.legend()
plt.show()
```
