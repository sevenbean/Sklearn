```python
digits=load_digits()
```

```python
digits.images.shape
```

```python
def showImage(V):
    plt.figure(figsize=(8,8))
    for i in range(20):
        plt.subplot(4,5,i+1)
        plt.imshow(V[i],cmap="binary")
        plt.xticks([])
        plt.yticks([])
```

```python
showImage(digits.images)
```

```python
'''
往原来的数据集添加噪音
normal里面的参数：loc 表示的是均值，scale表示的标准差，size表示的是数据矩阵的大小
'''
np.random.RandomState(42)
noisy=np.random.normal(digits.data,2)
```

```python
noisy.shape
```

```python
showImage(noisy.reshape(1797,8,8))
```

```python
#经过PCA降维之后的数据
from sklearn.decomposition import PCA
pca=PCA(n_components=0.8,svd_solver="full").fit(noisy)
```

```python
#获得SVD矩阵分解之后的特征矩阵
pca.components_.shape
```

```python
digits_inver=pca.transform(noisy)
digits_inver=pca.inverse_transform(digits_inver)
```

```python
digits_inver.shape
```

```python
showImage(digits_inver.reshape(1797,8,8))
```
